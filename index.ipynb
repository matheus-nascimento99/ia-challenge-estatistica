### Import libs

```python
import pandas as pd
```

### Invoicing dict

```python
dict_invoicing = {
   'date_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'value': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

### Dataframe creation

```python
df_invoicing = pd.DataFrame.from_dict(dict_invoicing)
df_invoicing
```

### Sales average

```python
sales_average = round(df_invoicing.value.mean(), 2)
sales_average
```

### Vertical bar chart

```python
df_invoicing.sort_values('date_ref').plot.bar(x='date_ref',y='value')
```

### Line chart

```python
line_chart = df_invoicing.plot('date_ref','value')
```
